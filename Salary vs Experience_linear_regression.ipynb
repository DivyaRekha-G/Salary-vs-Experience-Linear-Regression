```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
#Example Data

experience = np.array([1, 3, 5, 7, 10]).reshape(-1, 1)  # Independent variable (X)
salary = np.array([3, 6, 9, 12, 18])  # Dependent variable (Y) in LPA

```

```python
#Train model

model = LinearRegression()
model.fit(experience,salary)
```

```python
#predict

exp_test =np.array([[6]])  # predict salary for 6 years of experience
predict_salary =model.predict(exp_test)

print(f"Predicted salary for 6 years of experience:{predict_salary[0]}LPA")
```

```python
#Visalisation (plot)

plt.scatter(experience, salary, color="blue", label ="Actual Salaries")
plt.plot(experience,model.predict(experience),color="red",linewidth=2,label ="Regression Line")
plt.xlabel("Years of Experience")
plt.ylabel("Slary in (LPA)")
plt.legend()
plt.show()
```
